## **5. Choose a Prompting Technique**

Prompting is how you tell the model what to do.

**Examples of techniques:**

    * **Zero-shot prompting:** No examples, just direct instructions.
    * **Few-shot prompting:** Give 2‚Äì3 examples of input-output pairs so the model learns the pattern.
    * **Chain-of-thought prompting:** Ask the model to explain its reasoning step by step (for reasoning-heavy tasks).
    * **System prompt design:** Set the "persona" of the chatbot (friendly teacher, expert advisor, etc.).

**What to do:**

    * Clearly mention **which prompting technique(s) you used**.
    * Justify your choice ‚Äî why was this the best method for your use case?
    * Show a sample of your actual **prompt template** (system + user message).

### **Answer:**

#### **Techniques Used:**

- **System prompt design** (persona = friendly, supportive micro-coach).
- **Few-shot prompting** (examples of decision guidance included).
- **Chain-of-thought prompting** (for reasoning tasks like weighing pros/cons).

#### **Justification:**

- Few-shot ensures consistency in framing small decisions.
- Chain-of-thought avoids shallow answers, instead showing reasoning.
- System prompt keeps tone supportive, not judgmental.

#### **Sample Prompt Template:**

**System Prompt:**

Your role is to guide users through small everyday choices quickly and practically.

Greeting rule: Always begin the very first response with a friendly greeting such as 
'Hi, I‚Äôm your Micro-Coach. Tell me your situation and I‚Äôll help you think it through.'

Few-shot examples (follow this format):

Always be brief, practical, and supportive.
